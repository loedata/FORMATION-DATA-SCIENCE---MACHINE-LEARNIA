## Méthode 1 : PYPLOT

```python
# chargement librairie numpy
import numpy as np
```

```python
# chargement de librairie pyplot du package matplotlib
import matplotlib.pyplot as plt
```

###  Méthode : plot()

```python
#  on entre des données qu'on veut voir sur l'axe x et celles sur l'axe y
```

```python
# préparation de la fonction qu'on veut envoyer :
#  2 tableaux numpy 

# tableau x : comprend 10 points de 0 à 2 - x en abscisses
x = np.linspace(0, 2, 10)

# tableau y : comprend le carré de x - y en ordonnée
y = x**2

# définir le graphique ==f(x)
plt.plot(x,y)
# dessine le graphique
plt.show()
```

```python
# erreur : les tableaux doivent avoir les mêmes dimensions: 
# mêmes nbre de points
```

### Personnalisation de plot()

```python
# personnalisation de plot
# plot(x, y, label=..., lw=..., ls=..., c=...)
# label = nom de la courbe
# lw= épaisseur du trait
# ls = style du trait
# c = couleur du trait
```

### plt.plot(x,y, c=...) : couleur du trait

```python
#  changer la couleur du trait
plt.plot(x,y, c='green')
plt.show()
```

### plt.plot(x, y, lw=int...) : épaisseur du trait

```python
# changer l'épaisseur du trait
plt.plot(x,y, lw=5)
plt.show()
```

### plt.plot(x,y, ls='--') : style du trait

```python
# change le style du trait
plt.plot(x,y, ls='--')
plt.show()
```

```python
# on peut cumuler les personnalisations
plt.plot(x, y, lw=6, ls='--', c='green')
plt.show()
```

### plt.figure()

```python
# cycle de vie de la figure
# pour tracer un diagramme on commence par la feuille de travail
plt.figure()
# création d'un premier graphique
plt.plot(x, y)
plt.show()
```

### figsize : changer la taille de la figure

```python
# on peut changer la taille de la figure
plt.figure(figsize=(12, 8)) # en inches
plt.plot(x, y)
plt.show()
```

```python
# on peut ajouter d'autres figures
plt.figure() 
#  courbe 1
plt.plot(x, y)
# courbe 2
plt.plot(x, x**3)
plt.show()
```

### title : ajouter un titre à la figure

```python
# on peut ajouter un titre
plt.figure() 
plt.plot(x, y)
plt.plot(x, x**3)
plt.title('Figure 1')
plt.show()
```

### ajouter un nom aux axes : label=

```python
# on ajouter des noms aux axes
plt.figure() 
plt.plot(x, y)
plt.plot(x, x**3)
plt.title('Figure 1')
# axe des abscisses
plt.xlabel('Axe des abscisses')
# axe des ordonnées
plt.ylabel('Axe des ordonnées')
plt.show()
```

### ajouter une légende : légende + label

```python
# on peut ajouter une légende
plt.figure() 
plt.plot(x, y,label='Quadratique')
plt.plot(x, x**3, label='Cubique')
plt.title('Figure 1')
plt.xlabel('Axe des abscisses')
plt.ylabel('Axe des ordonnées')
plt.legend()
plt.show()
```

### plt.savefig() : sauvegarder la figure dans un fichier

```python
# sauvegarder la figure dans un fichier
# on peut ajouter une légende
plt.figure() 
plt.plot(x, y,label='Quadratique')
plt.plot(x, x**3, label='Cubique')
plt.title('Figure 1')
plt.xlabel('Axe des abscisses')
plt.ylabel('Axe des ordonnées')
plt.legend()
# à placer avant plt.show()
plt.savefig('figure.jpg')
plt.show()
```

```python
# créer une nouvelle figure continuer du code
plt.figure() 
plt.plot(x, y,label='Quadratique')
plt.plot(x, x**3, label='Cubique')
plt.title('Figure 1')
plt.xlabel('Axe des abscisses')
plt.ylabel('Axe des ordonnées')
plt.legend()
# à placer avant plt.show()
plt.savefig('figure.jpg')
plt.show()

# nouvelle figure...
plt.plot(x, y)
# ../..
plt.show()
```

### plt.sublpot() : affiche plusieurs graphiques dans même figure

### graphiques de haut en bas (vertical)

```python
# subplot() : grille de graphiques sur notre figure
# ex : 2 graphiques au sein d'une même figure
# premier graphique
plt.subplot(2,1,1)
plt.plot(x, y, c='red')
# premier graphique
plt.subplot(2,1,2)
plt.plot(x, y, c='blue')
plt.show()
```

### graphiques de droite à gauche (horizontal)

```python
# premier graphique
plt.subplot(2,2,1)
plt.plot(x, y, c='red')
plt.title('Graphique 1')
# second graphique
plt.subplot(2,2,2)
plt.plot(x, np.sin(x), c='green', label='Sinus')
plt.plot(x, np.cos(x), c='black', label='Cosinus')
plt.title('Graphique 2')
plt.legend()
plt.show()
```

## METHODE 2 : PROGRAMMATION ORIENTEE OBJET

```python
# OOP
# on créer des objets fig = figure objet, ax = axes tableau numpy ndarray
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
# création des graphiques avec mêmes abscisses
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(x, y)
ax[1].plot(x, np.sin(x))
plt.show() #une seule abscisse partagée
```

```python
# Un dictionnaires des données des 100 expériences
dataset={f'experience{i}': np.random.randn(100) for i in range(4)}
dataset
```

```python
# créer une fonction graphique qui imprime sur 1 même figure le dataset
#  4 expériences ==> 4 graphiques
def graphique(dataset):
    # taille du dictionnaire
    n = len(dataset)
    # figure agrandie dans lequel on va placer les 4 graphiques    
    plt.figure(figsize=(12,8))
    # boucle sur le dictionnaire
    for exp_cle, i in zip(dataset.keys(), range(1,n+1)):
        # Graphique i
        # n figure à l'emplacement i de 1 à 4       
        plt.subplot(n, 1, i)
        # alimente les ordonnées de l'expérience         
        plt.plot(dataset[exp_cle])
        # joute du titre = clé de l'expérience du dictionnaire
        plt.title(exp_cle)
    # tracer la figure    
    plt.show()
```

```python
graphique(dataset)
```
