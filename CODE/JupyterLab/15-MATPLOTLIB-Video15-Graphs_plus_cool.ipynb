```python
# chargement des librairies
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# dataset
iris = load_iris()
# print(iris)

# donn√©es
x = iris.data
# les targets classe 0 1 ou2
y = iris.target
names = list(iris.target_names)

print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables')
print(f'Il y a {np.unique(y).size} classes')
print('')
print(x)
print(y)
```

## (5) Graphique de CLASSIFICATION : plt.scatter

```python
# on fait passer en abscisse et ordonn√©es 2 des 4 variables 
# de notre probl√®me : par ex premi√®re variable et 2i√®me variable
# pour toutes les lignes
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

```python
# colorier ces points suivant la nature de leur classes y = 0 1 ou 2
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.xlabel('Longueur s√©pal')
plt.ylabel('Largeur du s√©pal')
plt.show()
```

```python
# limite : onne peut repr√©senter que 2 variables sur les 4
```

```python
# astuce : utiliser le param√®tre alpha pour contr√¥ler la transparence
# de chacun des points du graph.
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5)
plt.xlabel('Longueur s√©pal')
plt.ylabel('Largeur du s√©pal')
plt.show()
```

```python
# astuce : utiliser le param√®tre s qui contr√¥le la taille des points
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=100)
plt.xlabel('Longueur s√©pal')
plt.ylabel('Largeur du s√©pal')
plt.show()
```

```python
# astuce : controler la taille en fonction d'une variable du probl√®me
# ex : taille en fonction de la variable 3 x[:, 2]
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=x[:, 2]*100)
plt.xlabel('Longueur s√©pal')
plt.ylabel('Largeur du s√©pal')
plt.show()
```

```python
#  plus les points sont gros plus le p√©tal est long et inversement
```

## (4-2) surface : plot.surface()

```python
# visualiser des surfaces et fonctions maths mod√©lis√©es
# cr√©er une fonction f, g√©n√©rer par lambda g√©n√©rateur de fonction
# anonyme dans python
f = lambda x, y : np.sin(x) + np.cos(x+y)
# pour observer dans un graphique 3D :
# cr√©er 2 vecteurs numpy allant de 0 √† 5
# cr√©er une sorte de grille : np.meshgrid √† partir de 2 axes
X = np.linspace(0, 5, 100) # 100 points
Y = np.linspace(0, 5, 100)
# recr√©er X et Y
X, Y = np.meshgrid(X, Y)

# il reste √† injecter X, Y dans la fonction f  dans variable Z :
Z = f(X, Y)
```

```python
# dimension de Z : 100 lignes 100 colonnes
Z.shape
```

```python
# %matplotlib
ax = plt.axes(projection='3d')
# pr√©ciser la couleur : cmap
ax.plot_surface(X, Y, Z, cmap='plasma')
```

## (2) Contour plots : contour()

```python
# Repartir de la fonction f pour la visualiser en 2D  vue du dessus
# visualiser des surfaces et fonctions maths mod√©lis√©es
# cr√©er une fonction f, g√©n√©rer par lambda g√©n√©rateur de fonction
# anonyme dans python
f = lambda x, y : np.sin(x) + np.cos(x+y)
# pour observer dans un graphique 3D :
# cr√©er 2 vecteurs numpy allant de 0 √† 5
# cr√©er une sorte de grille : np.meshgrid √† partir de 2 axes
X = np.linspace(0, 5, 100) # 100 points
Y = np.linspace(0, 5, 100)
# recr√©er X et Y
X, Y = np.meshgrid(X, Y)

# il reste √† injecter X, Y dans la fonction f  dans variable Z :
Z = f(X, Y)

# visualisation en 2D en vue du dessus
plt.contour(X, Y, Z)
plt.show()
```

```python
# visualisation en 2D en vue du dessus
# pour √™tre plus visible : modifier le nombre de niveaux √† 20
# augmente la pr√©cision del'affichage
plt.contour(X, Y, Z, 20)
plt.show()
```

```python
# visualisation en 2D en vue du dessus
# pour √™tre plus visible : modifier le nombre de niveaux √† 20
# augmente la pr√©cision del'affichage
# encore plus lisible : modifier la couleur
plt.contour(X, Y, Z, 20, colors='black') # tous les positifs en continus,
# les n√©gatifs en discontinus
# on voit les maximum ou minimum si fonction co√ª
plt.show()
```

### avec une colormap : countourf(... cmap=)

```python
# visualisation en 2D en vue du dessus
# pour √™tre plus visible : modifier le nombre de niveaux √† 20
# augmente la pr√©cision del'affichage
# encore plus lisible : modifier la couleur avec colormap
plt.contourf(X, Y, Z, 20, cmap='RdGy') # tous les positifs en continus,
# les n√©gatifs en discontinus
# on voit les maximum ou minimum si fonction co√ªt
plt.show()
```

```python
# visualisation en 2D en vue du dessus
# pour √™tre plus visible : modifier le nombre de niveaux √† 20
# augmente la pr√©cision del'affichage
# encore plus lisible : modifier la couleur avec colormap
# encore plus : avec l√©gende
plt.contourf(X, Y, Z, 20, cmap='RdGy') # tous les positifs en continus,
# les n√©gatifs en discontinus
# on voit les maximum ou minimum si fonction co√ªt
plt.colorbar()
plt.show()
```

## CORRECTION EXERCICE VIDEO 14

```python
# Un dictionnaires des donn√©es des 100 exp√©riences
dataset={f'experience{i}': np.random.randn(100) for i in range(4)}
dataset

# cr√©er une fonction graphique qui imprime sur 1 m√™me figure le dataset
#  4 exp√©riences ==> 4 graphiques
def graphique(dataset):
    # taille du dictionnaire
    n = len(dataset)
    # figure agrandie dans lequel on va placer les 4 graphiques    
    plt.figure(figsize=(12,30))
    # boucle sur le dictionnaire
    for exp_cle, i in zip(dataset.keys(), range(1,n+1)):
        # Graphique i
        # n figure √† l'emplacement i de 1 √† 4       
        plt.subplot(n, 1, i)
        # alimente les ordonn√©es de l'exp√©rience         
        plt.plot(dataset[exp_cle])
        # joute du titre = cl√© de l'exp√©rience du dictionnaire
        plt.title(exp_cle)
    # tracer la figure    
    plt.show()
```

```python
graphique(dataset)
```

```python
# Un dictionnaires des donn√©es des 100 exp√©riences √† 6 classes
dataset_6={f'experience{i}': np.random.randn(100) for i in range(6)}
graphique(dataset_6) # 6 graphiques
```

```python
# Un dictionnaires des donn√©es des 100 exp√©riences √† 6 classes
# mais on collecte 3 variables
dataset_6_3={f'experience{i}': np.random.randn(100, 3) for i in range(6)}
graphique(dataset_6_3) # 6 graphiques
# chacune des 3 variables est affich√©e ur les graphiques
```
