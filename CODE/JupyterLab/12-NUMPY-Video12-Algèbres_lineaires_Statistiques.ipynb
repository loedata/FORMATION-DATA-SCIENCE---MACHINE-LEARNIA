### Fonction : np.unique()

```python
print(A)
# np.unique() : pour trouver le nombre de fois qu'un élément se répète
# à l'intérieur d'un tableau numpy
np.unique(A, return_counts=True)
# premier tableau : les éléments du tableau de gche à dte et haut vers bas
# deuxième tableau : le nombre de répétitions de l'élément
```

```python
# tableau plus grand
B = np.random.randint(0, 10, [5, 5])
B
```

```python
values, counts = np.unique(B, return_counts=True)
print(values)
print(counts)
```

```python
# utilisation de argsort pour classer le tableau pour trouver les
# éléments qui apparaissent le plus ou le moins
counts.argsort()
```

### Combiner np.unique() et argsort()

```python
# retrier le tableau values en fonction des index du tableau counts
values[counts.argsort()] # du nbre qui apparait le moins (6) et plus (3)
```

```python
# pour présenter les nombres du plus petit au plus grand
for i,j in zip(values[counts.argsort()], counts[counts.argsort()]):
    print(f'La valeur \'{i}\' apparait \'{j}\' fois')
```

## NaN Corrections

```python
# parfois il manque des données ou on a des erreurs ou sur des nombres
# il manque des données : NaN:Non A Number
# données corrompues
```

```python
# mean() sur un tableau qui comporte NaN ne fonctionne pas
A = np.random.rand(5, 5)
A[0, 2] = np.nan # force NaN ligne 1 col 3
A[4, 3] = np.nan # force NaN ligne 4 col 3
A
```

```python
A.mean() # moyenne ne fonctionne pas idem pour toutes les stats
```

```python
# numpy contient des fonctions pour faire la moyenne sans prendre les nan
```

### Routine : nanmean()

```python
# nanmean() : moyenne sans prendre en compte les nan
np.nanmean(A)
```

### Routine : nanvar()

```python
# nanvar() : variance sans prendre en compte les nan
np.nanvar(A)
```

### Routine : nanstd()

```python
# nanstd() : écart type sans prendre en compte les nan
np.nanstd(A)
```

###  Méthode : isnan()

```python
# compter le nombre de fois qu'on retrouve un nan dans le tableau
np.isnan(A) # mask numpy de boolean True = présence de nan
```

```python
# retouver le nombre ==> sum()
np.isnan(A).sum() # 2 entités nan
```

```python
# trouver le rapport de ces nan sur toutes les valeurs du tableau
np.isnan(A).sum()/A.size # 8 % des valeurs nana sur tous le tableau
```

```python
# supprimer les nan avec numpy grâce à boolean indexing
A[np.isnan(A)] = 0 # remplace les nan par la valeur par défaut 0 (ou 1...)
A
```

## EXERCICE

### Standardiser les différentes colonnes de la matrice A.

```python
# permet de mettre sur une même échelle toutes les colonnes, 
# lignes du dataset
```

```python
# plusieurs technique de standardisation :
```

```python
# Mise à l'échelle : (A-min(A)):(max(A)-min(A))
# Normalisation par la moyenne : (A-mean(A))/(max(A)-min(A))
# Très utile en machine learning : standardisation (A-mean(A))/std(A)
```

```python
# faire cette standardisation sur chaque colonne du tableau
# car les colonnes n'ont pas la même signification, prix, taille, surface
```

```python
# tableau de travail
import numpy as np
np.random.seed(0)
A = np.random.randint(0, 100, [10,5])
A
```

```python
D = (A - A.mean(axis=0)) / A.std(axis=0)
D
```

```python
# toutes les moyennes de D sont proches de 0 : la moyenne de chaque colonne
# =0
D.mean(axis=0)
```

```python
#  l'écart type de chaque colonne est égal à 1
D.std(axis=0)
```

```python
# chacune de nos colonnes suivent une distribution normale gaussienne
# absolument parfaite ==> ça facilite l'apprentissage statistique
```
