## Technique : Boolean Indexing

```python
# tableau rempli de nombres entiers aléatoires entre 0 et 10, dim 5*5
D = np.random.randint(0, 10, [5, 5])
D
```

```python
# on peut faire des opérations de comparaison

# tous les éléments inférieurs à 5
D < 5 # ==> tableau = un masque, on peut se servir comme un index à 
# l'intérieur de n'importe quel tableau numpy de même dimension
```

```python
# sélectionner tous les éléments de A<5 à remplacer par 10
D[D<5]=10
D
```

```python
# technique très utile pour convertir les valeurs d'un tableau
```

```python
# on peut cumuler les conditions
D[(D<8) & (D>5)] = 20
D
```

```python
# cette technique peut être utilisée our faire du traitement d'image
A = np.random.randint(0, 255, [1024, 720])
A
```

```python
# traitement pixel par pixel en modifiant A
# tous les pixels déjà luminosités élévés vont être égaux à 255
A[A>200] = 255
A
```

```python
# avec un masque,on peut filtrer les données d'un tableau de mêm dimension
# exemple : tableau pour lequel les éléments sont inférieurs à 5
D[D>10]
```

```python
# dimension du tableau est modifié 1 dimension
D[D>10].shape
```

```python
# photo de base dans scipy 
from scipy import misc
import matplotlib.pyplot as plt
# en noir et blanc
face = misc.face(gray=True)
# en noir et blanc
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```

```python
# dimension de la photo en noir et blanc : 2 dimensions 768 lignes 1024 col
face.shape
```

```python
# 1. zoomer de 1/4 vers le milieu
# indice : utiliser les dimensions de photos shape, enregistrer 
# ces dimensions dans ces variables et utiliser ces variables
# dans les techniques de slicing
h = face.shape[0]
w = face.shape[1]

face_zoom = face[h//4 : h//-4, w//4 : w//-4]
plt.imshow(face_zoom,cmap=plt.cm.gray)
plt.show()
```

```python
# augmenter la luminosité des pixels déjà au dessus d'une certaine 
# luminosité et réduire la luminosité des pixels déjà sombres
# via boolean indexing
face_clair=face
face_clair[face_clair>100]=255
face_clair[face_clair<100]=150
face_clair
plt.imshow(face_clair, cmap=plt.cm.gray)
plt.show()
```
